```python
import requests
import urllib.request
from bs4 import BeautifulSoup
```

```python
#first we inspect who site 
url = 'https://www.who.int/emergencies/diseases/novel-coronavirus-2019/situation-reports'
#with request library we get response from this site
response = requests.get(url)
```

```python
#then we parse this site as list of strings with BeautifulSoup library
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
#then we filter all the content by attributes of <a> tag (html) with build in function
page = soup.find_all(attrs={'target': '_blank'})
```

```python
#and list all reports from this page 
reports = [] 
for report in page:
    if 'Situation report' in str(report):
        reports.append(report)
```

```python
##we find last report simple, because it is first element of list 
#report does not have a link to who site so we concatinate each other
report_links = []
for report in reports:
    report_links.append(str(report.get('href'))) #BeatifulSoup function get()
print(len(report_links))
```

```python
#using urllib library we download pdf file of report from who site to our server
reports_num = len(reports)
report_name = ''
new_link = 'https://www.who.int'
for link in report_links:
    report_name = '/home/mykty-ayan/Data_Science_and_Machine_Learning/project/pdfs/report' + str(reports_num) + '.pdf'
    urllib.request.urlretrieve(new_link + link, report_name)
    reports_num -= 1
```
